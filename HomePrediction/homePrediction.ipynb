This notebook demonstrates how to build a **Machine Learning model** that predicts **housing prices in California** using **NumPy**, **Pandas**, and **Scikit-learn** within a **Jupyter Notebook**.

The primary objective is to predict the **median house value** based on **9 input (feature) variables**.

---

# Dataset Overview

The dataset contains **9 feature variables** and **1 target variable**.

## Feature Variables

| Feature | Description |
|----------|-------------|
| **longitude** | Measures how far west a house is located (higher values indicate farther west). |
| **latitude** | Measures how far north a house is located (higher values indicate farther north). |
| **housingMedianAge** | Median age of houses within a block (lower values indicate newer buildings). |
| **totalRooms** | Total number of rooms within a block. |
| **totalBedrooms** | Total number of bedrooms within a block. |
| **population** | Total population living within a block. |
| **households** | Total number of households (groups of people living together) within a block. |
| **medianIncome** | Median household income within a block (measured in tens of thousands of U.S. dollars). |
| **oceanProximity** | Describes how close the houses are to the ocean. |

---

## Target Variable

| Target | Description |
|---------|-------------|
| **medianHouseValue** | Median house value within a block (measured in U.S. dollars). |
